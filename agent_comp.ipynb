```python
from src.agents.DQN import DQN
from src.agents.DuelingDQN import DuelingDQN
from src.agents.A2C import Actor
import src.logger as logger
import src.agents.utils as AgentUtils
```

```python
import matplotlib.pyplot as plt

plt.plot(logger.process_average(dqn_rewards))
plt.plot(logger.process_average(dueling_rewards))
plt.plot(logger.process_average(actor_rewards))
plt.legend(['dqn', 'dueling', 'actor'])
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Comparison')
plt.show()
```
